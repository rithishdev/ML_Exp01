```python
import pandas as pd

df = pd.read_csv('/Housing.csv')

print("First 5 rows of the DataFrame:")
print(df.head())

print("\nColumn names and data types:")
df.info()

print("\nDescriptive statistics:")
print(df.describe())
```

```python
print("\nMissing values analysis:")
missing_values = df.isnull().sum()
missing_percentage = 100 * df.isnull().sum() / len(df)

missing_values_table = pd.DataFrame({
    'Missing Values': missing_values,
    '% of Total Values': missing_percentage
})

print(missing_values_table[missing_values_table['Missing Values'] > 0].sort_values(by='% of Total Values', ascending=False))

if missing_values_table['Missing Values'].sum() == 0:
    print("No missing values found in the dataset.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']

correlation_matrix = df[numerical_cols].corr()

print("\nCorrelation Matrix of Numerical Features:")
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()

key_numerical_features = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(key_numerical_features, 1):
    plt.subplot(2, 3, i) # Arrange plots in a 2x3 grid
    sns.scatterplot(x=df[feature], y=df['price'])
    plt.title(f'Price vs. {feature.replace("_", " ").title()}')
    plt.xlabel(feature.replace("_", " ").title())
    plt.ylabel('Price')

plt.tight_layout()
plt.show()
```
