```python
import pandas as pd

df = pd.read_csv('/content/scenario2.csv')


print("Dataset loaded successfully. Displaying the first 5 rows:")
df.head()
```

```python
print("\nDataFrame Information:")
df.info()

print("\nDescriptive Statistics:")
print(df.describe())

print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot histogram on the first subplot
sns.histplot(df['Glucose'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Glucose Levels (Histogram)')
axes[0].set_xlabel('Glucose Level')
axes[0].set_ylabel('Frequency')

# Plot boxplot on the second subplot
sns.boxplot(y=df['Glucose'], ax=axes[1], color='lightcoral')
axes[1].set_title('Distribution of Glucose Levels (Boxplot)')
axes[1].set_ylabel('Glucose Level')

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Display the plots
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots side-by-side for Age distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot histogram on the first subplot
sns.histplot(df['Age'], kde=True, ax=axes[0], color='lightgreen')
axes[0].set_title('Distribution of Age (Histogram)')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Plot boxplot on the second subplot
sns.boxplot(y=df['Age'], ax=axes[1], color='salmon')
axes[1].set_title('Distribution of Age (Boxplot)')
axes[1].set_ylabel('Age')

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Display the plots
plt.show()
```
